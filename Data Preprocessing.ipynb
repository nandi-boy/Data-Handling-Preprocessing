## Encodding

### One Hot Encodding

> One Hot Encoding is a method for converting categorical variables into a binary format. It creates new columns for each category where 1 means the category is present and 0 means it is not. The primary purpose of One Hot Encoding is to ensure that categorical data can be effectively used in machine learning models.

> A *dummy variable* (also known as indicator variable or just dummy) is one that takes a binary value (0 or 1) to indicate the absence or presence of some categorical effect that may be expected to shift the outcome.
> 
>  For example, if we were studying the relationship between biological sex and income, we could use a dummy variable to represent the sex of each individual in the study. The variable could take on a value of 1 for males and 0 for females (or vice versa).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Loan data - Copy.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
data["Gender"].fillna(data["Gender"].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
en_data=data[["Gender"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
obj = OneHotEncoder()
ar=obj.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=["Gender_female","Gender_male"])
```

```python
obj = OneHotEncoder(drop="first")  # drop="first" -> means drop the first category from each feature when encoding to avoid dummy variable trap (multicollinearity).
ar=obj.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=["Gender_male"])
```

### Ordinal Encodding

> Ordinal encoding assigns a unique integer to each category in a feature, reflecting their order. For example, in a dataset containing shirt sizes (small, medium, large) it can assign 1, 2 and 3 respectively. The point is not just to encode but to preserve the inherent ranking among categories. It is used in machine learning as various algorithms work best with numerical data only.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit_transform(df[["size"]])
```

```python
df["Size_en"] = oe.fit_transform(df[["size"]])
```

### Remove Outliers using Z Score

> One common technique for detecting outliers is Z score. It is a statistical measurement that describes how far a data point is from the mean, expressed in terms of standard deviations. It helps us to identify if a data point is relatively higher or lower than the mean and how far it deviates from the average value.

> Z= (X−μ)/σ 

> Where :
>  
> * X = Value of the data point
>
> * μ = Mean of the data
>
> * σ = Standard deviation of the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Loan data - Copy.csv")
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x ="age" ,data=data,color = "green")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x ="Principal" ,data=data,color = "green")
plt.show()
```

```python
sns.distplot(data["Principal"])
plt.show()
```

```python
sns.distplot(data["age"],color="green")
plt.show()
```

**Finding min_range and max_range to find outliers by 3rd standard deviation**

```python
min_range = data["age"].mean() - (3*data["age"].std())
max_range = data["age"].mean() + (3*data["age"].std())
```

```python
min_range,max_range
```

```python
new_dat = data[data["age"]<=max_range]
new_dat
```

```python
data.shape
```

```python
new_dat.shape
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x ="age" ,data=new_dat,color = "green")
plt.show()
```

```python
z_score = (data["age"]-data["age"].mean())/(data["age"].std())
```

```python
z_score
```

```python
data["Z_score"] = z_score
```

```python
data
```

```python
data[data["Z_score"]<3] 
```

## Function Transformer

> Function transformers are the type of feature transformation technique that uses a particular function to transform the data to the normal distribution.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Loan data - Copy.csv")
data.head(3)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
sns.boxplot(x=data["age"])
plt.show()
```

```python
sns.distplot(data["age"])
plt.show()
```

```python
q1 = data["age"].quantile(0.25)
q3 = data["age"].quantile(0.75)
IQR = q3-q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range,max_range
```

```python
data=data[data["age"]<=max_range]   # Replace the outlier free dataset for data["age"] with the original
```

```python
sns.boxplot(x=data["age"])
plt.show()
```

```python
sns.distplot(data["age"])
plt.show()
```

**Log Transform**

```python
from sklearn.preprocessing import FunctionTransformer
transform = FunctionTransformer(func=np.log1p)
transformed_data = transform.fit_transform(data["age"])
data["age_tfd"]=transformed_data
data.head(3)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["age"],color="green")

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["age_tfd"],color="orange")

plt.show()
```

**Square Transform**

```python
from sklearn.preprocessing import FunctionTransformer
transform = FunctionTransformer(func=lambda x : x**2)
transformed_data = transform.fit_transform(data["age"])
data["age_tfd_sq"]=transformed_data
data.head(3)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["age"],color="green")

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["age_tfd_sq"],color="orange")

plt.show()
```

## Handling Imbalanced Dataset

> **Random Oversampling:** Randomly duplicate examples in the minority class.
> 
> **Random Undersampling:** Randomly delete examples in the majority class.

```python
x = data.iloc[:,:-1]
y = data["Purchased"]
```

```python
from imblearn.over_sampling import RandomOverSampler
ru = RandomOverSampler()
ru_x,ru_y = ru.fit_resample(x,y)
print(ru_x)
print(ru_y)
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ru_x,ru_y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr1 = LogisticRegression()
lr1.fit(x_train,y_train)
```

```python
lr1.score(x_test,y_test)*100
```
